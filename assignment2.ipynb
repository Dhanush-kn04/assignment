1. Create the below pattern using nested for loop in Python.

```python
m1=int(input())
m2=m1*2
for row in range(0,m1):
    for col in range(0,row):
        print("*",end=" ")
    print("\n")
for row in range(m1+1,m2+1):
    for col in range (m2,row-1,-1):
        print("*", end=" ")
    print("\n")
```
